```python
hang_jung = (df["행정주소"] + " 주민센터").tolist()
```

```python
import requests

addrs = hang_jung
locations = []
coord = []
i = 0
for addr in addrs:
    try:
      url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(addr)
      headers = {"Authorization": "KakaoAK 35b8eaa0f5a9a86b8b4307a5465fe18c"}
      place = requests.get(url, headers = headers).json()["documents"]
      locations.append(place)
      coord.append([float(place[0]["x"]), float(place[0]["y"]), str(place[0]["road_address_name"])])
      print(i)
      i += 1
    except:
      coord.append([])
```
